Merging Both The Datasets

```python
df = pd.merge(df_csv,df_excel,on='LYLTY_CARD_NBR', how='left')
df.head()
```

```python
#dropping the duplicate values
df = df.drop_duplicates()
```

```python
#checking again for duplicated values
df.duplicated().sum()
```

```python
#checking the outliers
plt.boxplot(df['TOT_SALES'])
plt.show()
```

```python
#checking the date format
df = df[df['TOT_SALES'] > 0].copy()
df['DATE'] = pd.to_datetime(df['DATE'], errors='coerce')
```

```python
df['PACK_SIZE'] = df['PROD_NAME'].str.extract('(\d+)').astype(float)
df['BRAND'] = df['PROD_NAME'].str.split().str[0]
```
