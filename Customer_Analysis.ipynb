```python
#checking the date format
df = df[df['TOT_SALES'] > 0].copy()
df['DATE'] = pd.to_datetime(df['DATE'], errors='coerce')
```

```python
df['PACK_SIZE'] = df['PROD_NAME'].str.extract('(\d+)').astype(float)
df['BRAND'] = df['PROD_NAME'].str.split().str[0]
```

```python
#DATA + YEARMONTH
df['DATE'] = pd.to_datetime(df['DATE'], errors='coerce')
df['YEARMONTH'] = df['DATE'].dt.strftime('%Y%m').astype(int)
```

```python
#CREATE METRICS
measureOverTime = df.groupby(['STORE_NBR', 'YEARMONTH']).agg(
    totSales=('TOT_SALES', 'sum'),
    nCustomers=('LYLTY_CARD_NBR', 'nunique'),
    nTxn=('TXN_ID', 'nunique'),
    totalQty=('PROD_QTY', 'sum')
).reset_index()

# Derived metrics
measureOverTime['nTxnPerCust'] = measureOverTime['nTxn'] / measureOverTime['nCustomers']
measureOverTime['nChipsPerTxn'] = measureOverTime['totalQty'] / measureOverTime['nTxn']
measureOverTime['avgPricePerUnit'] = measureOverTime['totSales'] / measureOverTime['totalQty']
```

```python
#FILTERING PRE TRIALS
preTrialMeasures = measureOverTime[measureOverTime['YEARMONTH'] < 201902]
```

```python
#USING CORRELATION FUNCTIONS
def calculate_correlation(df, metric, trial_store):
    stores = df['STORE_NBR'].unique()
    results = []

    for store in stores:
        if store != trial_store:
            merged = pd.merge(
                df[df['STORE_NBR'] == trial_store][['YEARMONTH', metric]],
                df[df['STORE_NBR'] == store][['YEARMONTH', metric]],
                on='YEARMONTH',
                suffixes=('_trial', '_control')
            )

            corr = merged[f"{metric}_trial"].corr(merged[f"{metric}_control"])
            results.append([trial_store, store, corr])

    return pd.DataFrame(results, columns=['Store1', 'Store2', 'corr'])
```

```python
#MAGNITUDE FUNCTIONS
def calculate_magnitude(df, metric, trial_store):
    stores = df['STORE_NBR'].unique()
    results = []

    for store in stores:
        if store != trial_store:
            merged = pd.merge(
                df[df['STORE_NBR'] == trial_store][['YEARMONTH', metric]],
                df[df['STORE_NBR'] == store][['YEARMONTH', metric]],
                on='YEARMONTH',
                suffixes=('_trial', '_control')
            )

            merged['diff'] = abs(merged[f"{metric}_trial"] - merged[f"{metric}_control"])

            score = 1 - (merged['diff'] - merged['diff'].min()) / (merged['diff'].max() - merged['diff'].min())
            results.append([trial_store, store, score.mean()])

    return pd.DataFrame(results, columns=['Store1', 'Store2', 'mag'])
```

```python
#FUNCTION TO GET CONTROL STORE
def get_control_store(trial_store):
    corr_sales = calculate_correlation(preTrialMeasures, 'totSales', trial_store)
    corr_cust = calculate_correlation(preTrialMeasures, 'nCustomers', trial_store)

    mag_sales = calculate_magnitude(preTrialMeasures, 'totSales', trial_store)
    mag_cust = calculate_magnitude(preTrialMeasures, 'nCustomers', trial_store)

    score_sales = pd.merge(corr_sales, mag_sales, on=['Store1','Store2'])
    score_sales['score_sales'] = 0.5 * score_sales['corr'] + 0.5 * score_sales['mag']

    score_cust = pd.merge(corr_cust, mag_cust, on=['Store1','Store2'])
    score_cust['score_cust'] = 0.5 * score_cust['corr'] + 0.5 * score_cust['mag']

    final = pd.merge(score_sales[['Store1','Store2','score_sales']],
                     score_cust[['Store1','Store2','score_cust']],
                     on=['Store1','Store2'])

    final['finalScore'] = (final['score_sales'] + final['score_cust']) / 2

    return final.sort_values('finalScore', ascending=False).iloc[1]['Store2']
```

```python
#FIND CONTROL STORES
control_77 = get_control_store(77)
control_86 = get_control_store(86)
control_88 = get_control_store(88)

print("Control Store for 77:", control_77)
print("Control Store for 86:", control_86)
print("Control Store for 88:", control_88)
```

```python
#SCALING + % DIFFERENCE
def trial_analysis(trial_store, control_store):
    trial = measureOverTime[measureOverTime['STORE_NBR'] == trial_store]
    control = measureOverTime[measureOverTime['STORE_NBR'] == control_store]

    # Scaling factor
    scaling = trial[trial['YEARMONTH'] < 201902]['totSales'].sum() / \
              control[control['YEARMONTH'] < 201902]['totSales'].sum()

    control['scaledSales'] = control['totSales'] * scaling

    merged = pd.merge(
        trial[['YEARMONTH','totSales']],
        control[['YEARMONTH','scaledSales']],
        on='YEARMONTH'
    )

    merged['pctDiff'] = abs(merged['totSales'] - merged['scaledSales']) / merged['scaledSales']

    return merged
```

```python
#RUN ANALYSIS
analysis_77 = trial_analysis(77, control_77)
analysis_86 = trial_analysis(86, control_86)
analysis_88 = trial_analysis(88, control_88)
```

```python
#VISUALIZATION
def plot_trial(trial_store, control_store):
    trial = measureOverTime[measureOverTime['STORE_NBR'] == trial_store]
    control = measureOverTime[measureOverTime['STORE_NBR'] == control_store]

    plt.figure(figsize=(10,5))
    plt.plot(trial['YEARMONTH'], trial['totSales'], label='Trial')
    plt.plot(control['YEARMONTH'], control['totSales'], label='Control')

    plt.axvspan(201902, 201904, color='grey', alpha=0.2)  # trial period

    plt.title(f"Store {trial_store} vs Control {control_store}")
    plt.legend()
    plt.show()
```

```python
plot_trial(77, control_77)
plot_trial(86, control_86)
plot_trial(88, control_88)
```
